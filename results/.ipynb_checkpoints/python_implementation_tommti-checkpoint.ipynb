```python
##### read bitops csv
def parse_benchmark(filename="test_tommti20211117T103120.csv",microunits=True):
    data= pd.read_csv(filename,sep=";")
    if microunits : 
        data=data.applymap(lambda x : x/ 1e6 if  type(x) == int or type(x) == float  else x) # convert units from milli to standard
    data[["env","benchmark","iteration"]]= data["NAME"].str.split("_",expand=True)
    data["version"]=data["env"].apply(lambda x:version[x])
    data["category"]=data["env"].apply(lambda x :category[x] )
    return data
data=parse_benchmark()
data.head()
```

```python
benchmarks=['intArithmetic', 'doubleArithmetic', 'longArithmetic', 'trig',
       'array', 'hashes', 'heapsort', 'vector', 'matrixMultiply',
       'nestedLoop', 'stringConcat', 'io', 'except']

measures=["WARMUP_CPU_0","WARMUP_CPU_1","WARMUP_DRAM_0","WARMUP_DRAM_1","WARMUP_DURATION_","EXECUTION_CPU_0",
          "EXECUTION_CPU_1","EXECUTION_DRAM_0","EXECUTION_DRAM_1","EXECUTION_DURATION"]
benchmark=benchmarks[1]
measure="EXECUTION_CPU_1"
```

```python
def add_line(ax, xpos):
    line = plt.Line2D([xpos, xpos], [1,-0.3],
                      transform=ax.transAxes, color='black')
    line.set_clip_on(False)
    ax.add_line(line)

def get_ratio(data)  :
    separator=data.groupby("version").count()
    separator=separator.loc[3][0]/(separator.loc[3][0]+separator.loc[2][0])
    return separator
    
def plot_box(df,measure="EXECUTION_CPU_1"):
    fig, axes = plt.subplots(1, 1, figsize=(32, 14))
 
    fig=df.boxplot(by="env",column=measure,ax=axes,rot=-90,return_type="both",showfliers=False)
    fig[0].ax.set_xticklabels(list(df["env"].unique()))
    fig[0].ax.set_ylabel(ylabel="Energy Consumption (J)",labelpad=22,fontsize="30")
    fig[0].ax.get_figure().suptitle('')
    fig[0].ax.set_xlabel(xlabel='')
    
    for i in range(len(fig[0][1]['boxes'])): 
        color = pycolors[fig[0][0].get_xticklabels()[i].get_text()]
        fig[0][1]['boxes'][i].set(color="b",edgecolor='black',facecolor=color,linewidth=1)
        fig[0][1]['medians'][i].set(color="red",linewidth=0.5)
        fig[0][1]['whiskers'][i].set(color="black",linewidth=1)
    plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)
    return fig,axes


def plot_bar(data,measure="EXECUTION_CPU_1"):
    fig, axes = plt.subplots(1, 1, figsize=(32, 14))
    df = data.groupby("env").mean().reset_index()
    df["position"]=df["env"].apply(lambda x: envs.index(x) )
    mini = min(df["position"])
    df["position"]=df["position"].apply(lambda x : x- mini+1)
    df.sort_values("position",inplace=True)
    palette=[pycolors[e] for e in df["env"] ]
    fig=df.plot.bar(x="env",y=measure,ax=axes,rot=-90,color=palette)
    separator=get_ratio(df)

    add_line(fig,separator )
    fig.set_ylabel(ylabel="Energy Consumption (J)",labelpad=22,fontsize="30")
    fig.get_figure().suptitle('')
    fig.set_xlabel(xlabel='')
    fig.get_legend().remove()
    axes.grid("True" )
    fig.text(x=0.5*separator*(fig.get_xlim()[1]-fig.get_xlim()[0]),y=-0.3*fig.get_ylim()[1],s='python3',ha="right",size=30)
    fig.text(x=1.5*separator*(fig.get_xlim()[1]-fig.get_xlim()[0]),y=-0.3*fig.get_ylim()[1],s='python2',ha="right",size=30)

    #     fig.annotate('python2',xy=(0.25,300),size=25)
    return fig,axes
#     fig[0].ax.legend(handles=legends, loc='lower left',fontsize='xx-large',labelspacing=1)
fig,x=plot_bar(data.loc[data["benchmark"]==benchmark])
```

```python
for bench in benchmarks:
    fig,x=plot_bar(data.loc[data["benchmark"]==bench])
#     fig.set_title(bench,fontsize="30",pad=50 )
    plt.savefig(f"bar_plot_tommti{bench}.pdf")
```

```python
data2=data.groupby(["env","benchmark"]).mean().reset_index()
import math
l=[]
for benchmark in data2["benchmark"].unique():
    df = data2.loc[data2["benchmark"]==benchmark].sort_values(measure,ignore_index=True)
    df["rank"]= df.index
    mini=math.log(min(df[measure]))
    df["performance"]=df[measure].apply(lambda x : mini/math.log(x) *10)
    l.append(df)

data2=pd.concat(l)
# data2["bench_order"]=data2["benchmark"].apply(lambda x : benchs_orde.index(x))
# data2=data2.sort_values(["bench_order","rank"])
data2.fillna(0)
```

```python
df= data2.loc[:,["env","benchmark","performance"]]
df=df.pivot(index="env",columns=["benchmark"],values=["performance"]).droplevel(level=0,axis=1).reset_index()
df=df.replace(np.NaN, 0)
# df=df.rename(lambda x : benchs_orde[x],axis=1).reset_index()
show(df)
```

```python
from math import pi


def plot_radar(df,line,ax=None):

    categories=list(df)[1:]
    N = len(categories)
    label=df.loc[line]["env"]
     # We are going to plot the first line of the data frame.
    # But we need to repeat the first value to close the circular graph:
    values=df.loc[line].drop('env').values.flatten().tolist()
    values += values[:1]
 
    # What will be the angle of each axis in the plot? (we divide the plot / number of variable)
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]
 
    # Initialise the spider plot
    if ax is None :
        ax = plt.add_subplot(111, projection='polar')
 
    # Draw one axe per variable + add labels
    plt.xticks(angles[:-1], categories)
    plt.yticks([2,4,6,8,10], ["1","10","100","5000","1000"], color="grey")
    # Draw ylabels
    ax.set_rlabel_position(0)
    ax.tick_params(axis='x', which='major', pad=30)
    
    plt.ylim(0,12)
 
    # Plot data
    ax.plot(angles, values, linewidth=1,color=pycolors[label], linestyle='solid')
    # Fill area
    ax.fill(angles, values, color=pycolors[label], alpha=0.5,label=label)
#     plt.legend(loc='upper right', bbox_to_anchor=(1, 1))

    return ax
    # Show the graph

```

```python
## generic figure 
fig=plt.figure(figsize=(30, 50))
row=12
for row in range(len(df)):
    ax=fig.add_subplot(5,3,row+1, projection='polar')
    plot_radar(df,row,ax)
    label=df.loc[row]["env"]
    ax.set_title(label,color=pycolors[label],size=22,y=1.15)
plt.tight_layout()
plt.savefig("alltomti_performance.pdf")
```

```python
def plot_radar_comparaison(df,*rows):

    fig=plt.figure(figsize=(14, 14))
    ax=fig.add_subplot(111, projection='polar')
    label=""
    for row in rows:
        plot_radar(df,row,ax)
        label += "_"+df.loc[row]["env"]
        
    plt.legend(loc='upper right', bbox_to_anchor=(1.1, 0),prop={'size': 30})
    plt.savefig(f"tommti_compare_{label}.pdf")
```

```python
plot_radar_comparaison(df,12,13)# pypy2 vs pypy3  
plot_radar_comparaison(df,10,11) numba 2 3 
plot_radar_comparaison(df,4,5) # intelpython 2 ,3 
plot_radar_comparaison(df,1,2) #cpython 2 3 
plot_radar_comparaison(df,1,12,10)

plot_radar_comparaison(df,2,11,13)
```
